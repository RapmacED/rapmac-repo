```python
def hrefClean(href):
    href = str(href).replace('?trk=public_profile_experience-item_profile-section-card_subtitle-click','')
    return href

def cleanDate(date,duration):
    date = date.replace(duration,'')
    return date
```

```python
df = pd.read_csv('/Users/raph/Documents/Dataship_Code/urlsLinkedinTest.csv')
title_main = []
subtitle_main = []
title = []
subtitle = []
link = []
date = []
duration = []
description = []
for i in range(len(df)):
    driver=webdriver.Chrome()
    url = df['url'][i]
    driver.get(url)
    time.sleep(3)
    source = driver.page_source
    soup = BeautifulSoup(source)
    title_main.append(soup.find("h1", {"class":"top-card-layout__title font-sans text-lg papabear:text-xl font-bold leading-open text-color-text mb-0"}).getText().strip())
    subtitle_main.append(soup.find("h2",{"class": 'top-card-layout__headline break-words font-sans text-md leading-open text-color-text'}).getText().strip())
    title_sub = []
    subtitle_sub = []
    link_sub = []
    date_sub = []
    duration_sub = []
    description_sub = []
    soupTitle = soup.findAll("h3",{"class": 'profile-section-card__title'})
    soupSubtitle = soup.findAll("h4",{"class": 'profile-section-card__subtitle'})
    soupMeta = soup.findAll("div",{"class": 'profile-section-card__meta'})
    count=0
    for s in soupMeta:
        childrenDate = s.findChildren('span')
        childrenDuration = s.findChildren('span',{"class": 'date-range__duration'})
        childrenDescription = s.findChildren('p', {"class": 'show-more-less-text__text--less'})
        title_sub.append(soupTitle[count].getText().strip())
        subtitle_sub.append(soupSubtitle[count].getText().strip())
        for s in soupSubtitle[count]:
            href = False
            if str(type(s))== "<class 'bs4.element.Tag'>":
                link_sub.append(hrefClean(s.get('href')))
                href = True
        if href == False:
            link_sub.append('')
        if childrenDuration != []:
            date_sub.append(cleanDate(childrenDate[0].getText().strip(),childrenDuration[0].getText().strip()))
            duration_sub.append(childrenDuration[0].getText().strip())
        elif childrenDate != []:
            date_sub.append(childrenDate[0].getText().strip() )
            duration_sub.append('')
        else:
            date_sub.append('')
            duration_sub.append('')
        if childrenDescription != []:
            description_sub.append(childrenDescription[0].getText().strip())
        else:
            description_sub.append('')
        count+=1
    title.append(title_sub)
    subtitle.append(subtitle_sub)
    link.append(link_sub)
    date.append(date_sub)
    duration.append(duration_sub)
    description.append(description_sub)
    driver.quit()
```
