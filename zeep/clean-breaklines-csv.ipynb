```python
import csv

with open('/Users/raph/Desktop/AirTable.csv', "r") as readfile:

    name_split = []
    
    for line in readfile:
        whitespace_split = line.split(" ")

        remove_returns = (line.replace('\n', "") for line in whitespace_split)

        name_split.append(remove_returns)

    print (name_split)
```

```python
with open ('/Users/raph/Desktop/AirTableCleaned.csv', 'w', newline = '\n')     as csvfile:

        writer = csv.writer(csvfile, delimiter = ';',
                            quotechar = '"', quoting = csv.QUOTE_MINIMAL)

        writer.writerows(name_split)
```
